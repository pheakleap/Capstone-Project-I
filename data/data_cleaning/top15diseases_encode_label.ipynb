```python
all_symptoms
```

```python
le = LabelEncoder()
```

```python
all_symptoms = sorted(all_symptoms)  
le.fit(all_symptoms)
```

```python
for col in df.columns:
    if 'Symptom' in col:
        df[col] = df[col].map(lambda x: le.transform([x])[0])
```

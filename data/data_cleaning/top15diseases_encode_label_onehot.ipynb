```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('../processed/Top15diseases_clean.csv')
```

```python
df.drop_duplicates('inplace=True')
```

```python
Symptom_1 = set(df['Symptom_1'].unique())
Symptom_2 = set(df['Symptom_2'].unique())
Symptom_3 = set(df['Symptom_3'].unique())
Symptom_4 = set(df['Symptom_4'].unique())
Symptom_5 = set(df['Symptom_5'].unique())
Symptom_6 = set(df['Symptom_6'].unique())
Symptom_7 = set(df['Symptom_7'].unique())
Symptom_8 = set(df['Symptom_8'].unique())
Symptom_9 = set(df['Symptom_9'].unique())
Symptom_10 = set(df['Symptom_10'].unique())
Symptom_11 = set(df['Symptom_11'].unique())
Symptom_12 = set(df['Symptom_12'].unique())
Symptom_13 = set(df['Symptom_13'].unique())
Symptom_14 = set(df['Symptom_14'].unique())
Symptom_15 = set(df['Symptom_15'].unique())
Symptom_16 = set(df['Symptom_16'].unique())
Symptom_17 = set(df['Symptom_17'].unique())
print(Symptom_1.union(Symptom_2, Symptom_3, Symptom_4, Symptom_5, Symptom_6, Symptom_7, Symptom_8, Symptom_9, Symptom_10, Symptom_11, Symptom_12, Symptom_13, Symptom_14, Symptom_15, Symptom_16, Symptom_17))
all_symptoms = Symptom_1.union(Symptom_2, Symptom_3, Symptom_4, Symptom_5, Symptom_6, Symptom_7, Symptom_8, Symptom_9, Symptom_10, Symptom_11, Symptom_12, Symptom_13, Symptom_14, Symptom_15, Symptom_16, Symptom_17)
```

```python
le = LabelEncoder()
```

```python
all_symptoms = sorted(all_symptoms)  
le.fit(all_symptoms)
```

```python
for col in df.columns:
    if 'Symptom' in col:
        df[col] = df[col].map(lambda x: le.transform([x])[0])
```

```python
print("Before balancing:", Counter(df["Disease"]))
```

```python
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[['Disease']])

one_hot_df = pd.DataFrame(one_hot_encoded, 
                          columns=encoder.get_feature_names_out(["Disease"]))

df_encode = pd.concat([df.drop(columns=["Disease"], axis=1), one_hot_df], axis=1)
```

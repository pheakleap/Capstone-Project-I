```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../data/processed/Cleaned_Top15Diseases.csv")
```

```python
df.head()
```

Get unique symptom

```python
disease_col = 'Disease'
symptom_cols = ['Symptom_1', 'Symptom_2', 'Symptom_3',
                'Symptom_4', 'Symptom_5', 'Symptom_6',
                'Symptom_7', 'Symptom_8', 'Symptom_9',
                'Symptom_10', 'Symptom_11', 'Symptom_12',
                'Symptom_13', 'Symptom_14', 'Symptom_15',
                'Symptom_16', 'Symptom_17']
```

```python
unique_symptoms = set()
feature = df.drop(disease_col, axis=1)
for col in feature.columns:
    unique_symptoms.update(feature[col].unique())

symptom_mapping = {symptom: i for i, symptom in enumerate(sorted(unique_symptoms))}

for col in symptom_cols:
    df[col] = df[col].map(symptom_mapping)
```

Count all the symptoms that have in top 15 diseases

```python
import pandas as pd
```

```python
#Load the data
file_path = '../data/Cleaned_Top15Diseases.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#Count Unique symptoms names
symptoms_columns = ['Symptom_1', 'Symptom_2', 'Symptom_3', 
                    'Symptom_4', 'Symptom_5', 'Symptom_6',
                    'Symptom_7', 'Symptom_8', 'Symptom_9',
                    'Symptom_10', 'Symptom_11', 'Symptom_12',
                    'Symptom_13', 'Symptom_14', 'Symptom_15',
                    'Symptom_16', 'Symptom_17']
#Get the unique symptoms names
unique_symptoms = set()
for col in symptoms_columns:
    unique_symptoms.update(df[col].dropna().unique())

#Remove "None" if it exists
unique_symptoms.discard("None")
unique_symptoms.discard('enlarged_thyroid')
unique_symptoms = sorted(unique_symptoms)

print(f"Unique symptoms: ({len(unique_symptoms)}): ")
display(unique_symptoms)
```

```python
# Load the original Top15diseases.csv file
df_original = pd.read_csv(file_path)

# Verify the unique diseases in the original file
unique_diseases_original = df_original['Disease'].unique()
print(f'Unique diseases in original file ({len(unique_diseases_original)}): {unique_diseases_original}')
```

# Cleaning the top 15 diseases

```python
#Import necessary libraries
import pandas as pd
```

```python
#Load the datasets
file_path = "../data/Top15Diseases.csv"
diseases_df = pd.read_csv(file_path)
diseases_df.head()
```

```python
# Count the occurrences of each disease in the dataset
disease_counts = diseases_df['Disease'].value_counts()
print(f'Disease counts:\n{disease_counts}')
```

```python
# Fill NaN values with "None"
diseases_df.fillna("None", inplace=True)

# Combine symptom columns into a single Symptoms column
symptom_columns = [col for col in diseases_df.columns if "Symptom" in col and col != "Symptom_Count"]
diseases_df["Symptoms"] = diseases_df[symptom_columns].values.tolist()

# Clean the Symptoms column
diseases_df["Symptoms"] = diseases_df["Symptoms"].apply(lambda x: [s for s in x if s != "None"])

# Count the occurrences of each disease in the cleaned dataset
disease_counts_cleaned = diseases_df['Disease'].value_counts()
print(f'Disease counts after cleaning:\n{disease_counts_cleaned}')
```

```python
diseases_df.head()
```

```python
# Save the cleaned dataset to a new CSV file
output_file_path = '../data/Cleaned_Top15Diseases.csv'
try:
    diseases_df.to_csv(output_file_path, index=False)
    print(f"File saved successfully to {output_file_path}")
except Exception as e:
    print(f"An error occurred while saving the file: {e}")
```

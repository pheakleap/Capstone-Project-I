Label Encodings for changes to numeric values

```python
# Import necessary libraries
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
import ast
```

```python
# Load the data from the CSV file
file_path = '../data/Cleaned_Top15Diseases.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Count the occurrences of each disease in the dataset
disease_counts = df['Disease'].value_counts()
print(f'Disease counts:\n{disease_counts}')
```

```python
# Drop unnecessary columns
df = df.drop(columns=['Symptom_Count', 'Symptoms'])
```

```python
# Initialize Label Encoder
label_encoder = LabelEncoder()
```

```python
# Encode 'Symptom_1' to 'Symptom_17', replacing None values with 0 before encoding
symptom_columns = [f'Symptom_{i}' for i in range(1, 18)]
for col in symptom_columns:
    df[col] = df[col].fillna('None')  # Replace NaN with 'None'
    df[col] = label_encoder.fit_transform(df[col])  # Convert categorical to numerical
```

```python
# Save the processed dataset
df.to_csv("../data/Encoded_Cleaned_Top15Diseases.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import json
```

```python
df = pd.read_csv("D:/Term7/Capstone-Project-I/data/processed/Cleaned_Top15Diseases.csv") 
df.head()
```

```python
df = df.drop(columns=['Symptom_Count', 'Symptoms'])
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.fillna('none', inplace=True)
df.head()
```

```python
symptom_df = df.drop(columns='Disease')
disease_df = df['Disease']
```

```python
all_symptoms = set().union(*[set(symptom_df[col].unique()) for col in symptom_df.columns if col.startswith('Symptom_')])
all_symptoms = sorted(all_symptoms)
print(all_symptoms)
```

```python
label_encoder = LabelEncoder()
all_symptoms_encoded = label_encoder.fit_transform(all_symptoms)
```

```python
for col in df.columns:
    if 'Symptom' in col:
        df[col] = df[col].map(lambda x: label_encoder.transform([x])[0])
```

```python
all_symptoms_encode = {symptom: int(encode) for symptom, encode in zip(all_symptoms, all_symptoms_encoded)}
```

```python
all_diseases = disease_df.unique()
all_diseases = sorted(all_diseases)
print(all_diseases)
```

```python
all_disease_encoded = label_encoder.fit_transform(all_diseases)
```

```python
all_disease_encode = {disease: int(encode) for disease, encode in zip(all_diseases, all_disease_encoded)}
```

```python
# Encode Disease column (if it exists)
if 'Disease' in df.columns:
    # Fit the encoder on the entire column, then transform
    df['Disease'] = label_encoder.fit_transform(df['Disease'])
```

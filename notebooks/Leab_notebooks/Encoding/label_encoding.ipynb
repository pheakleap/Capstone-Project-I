Label Encodings for changes to numeric values

```python
#Import neccesary libraries
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
```

```python
#Load the data from the CSV files
file_path = 'D:\CADT University\CADT-Y3\CodeAllSub\Capstone\Capstone-Project-I\data\processed\lyleab_file\Cleaned_Top15Diseases.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Count the occurrences of each disease in the dataset
disease_counts = df['Disease'].value_counts()
print(f'Disease counts:\n{disease_counts}')
```

```python
#Drop all columns in dataset except "Disease", "Symptom_Count" and "Symptoms" columns
df_filltered = df[["Disease", "Symptom_Count", "Symptoms"]]
df_filltered.head()
```

```python
#Encode the "Disease" column
label_encoder_disease = LabelEncoder()
df_filltered["Disease_Encoded"] = label_encoder_disease.fit_transform(df_filltered['Disease'])
df_filltered.head()
```

```python
#Create the mapping of encoded values to original values
disease_mapping = dict(zip(label_encoder_disease.classes_, label_encoder_disease.transform(label_encoder_disease.classes_)))
disease_mapping
```

```python
#Count the uniques symptomms names in the "Symptoms" columns
unique_symptoms = set()
for symptoms_list in df_filltered["Symptoms"]:
    symptoms = symptoms_list.strip("[]").replace("'", "").split(", ")
    unique_symptoms.update(symptoms)
len(unique_symptoms)
```

```python
#Create a mapping of symptoms before encode values
label_encoder_symptoms = LabelEncoder()
label_encoder_symptoms.fit(list(unique_symptoms))
symptom_mapping = dict(zip(label_encoder_symptoms.classes_, label_encoder_symptoms.fit_transform(label_encoder_symptoms.classes_)))
symptom_mapping
```

```python
#Encode each symptoms in the "Symptoms" columns using the label_encoder_symptoms
df_filltered["Symptom_Encoded"] = df_filltered['Symptoms'].apply(lambda x: [symptom_mapping[symptom] for symptom in x.strip("[]").replace("'", "").split(", ")])
df_filltered.head()
```

```python
#Save the encoded data for training and testing models
encoded_file_path = 'D:\CADT University\CADT-Y3\CodeAllSub\Capstone\Capstone-Project-I\data\processed\lyleab_file\Encoded_Cleaned_Top15Diseases.csv'
df_filltered.to_csv(encoded_file_path, index=True)
```

```python
le = LabelEncoder()
```

```python
all_symptoms = df.filter(like="Symptom").values.flatten()  # Get all unique symptom values
all_symptoms = list(set(map(str, all_symptoms)))  # Convert everything to string and get unique values

le.fit(all_symptoms)  # Fit on all symptoms before transforming
```

```python
for col in df.columns:
    if 'Symptom' in col:
        df[col] = df[col].map(lambda x: le.transform([x])[0])
```
